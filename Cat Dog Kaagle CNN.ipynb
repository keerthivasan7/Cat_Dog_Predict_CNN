```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import zipfile
from zipfile import ZipFile
import os
import cv2
import shutil
from tensorflow import keras
from tensorflow.keras import layers
```

```
gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255.0, validation_split=0.2)
```

```
dataframe = pd.DataFrame(columns=["filename","label"])
dataframe['filename'] = os.listdir('/databricks/driver/train/')
dataframe['label'] = [i.split(".")[0] for i in os.listdir('/databricks/driver/train/')]
```

```
train_gen = gen.flow_from_dataframe(dataframe,"/databricks/driver/train/",x_col='filename',y_col='label',target_size=(150,150), batch_size=32, subset='training')
```

```
#Notes:
#only 20000 images is there but there are 250000, In generator we mentioned we need validation data also
# if need to generate same random values in batching we can use seed
```

```
val_gen = gen.flow_from_dataframe(dataframe,"/databricks/driver/train/",x_col='filename',y_col='label',target_size=(150,150), batch_size=32, subset='validation')
```

### CNN

```
#del model
```

```
model = tf.keras.models.Sequential()
model.add(tf.keras.Input((150,150,3)))
#chunk
model.add(tf.keras.layers.Conv2D(filters=16 ,kernel_size=(3,3), strides=(1,1), padding='valid', activation='relu'))
#kernal_size is the filtered image size =>[random size matrix] which got multiplied with image pixel
#conv2d always coverts to 3 channel(RGB) to 1 channel (grayscale)
#filter is the random for conv2d multiply
#16 features for single image like (sharpen, box etc)
# Formula for Feature 

# Feature = (image - kernel + stride) / stride 
# if this is not fully divisible please add padding 

# Feature = (image - kernal + stride + padding) / stride 
#formula => 150 -3 +1 /1 => 148 size after conv2D
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2), strides=None))
#formula => 148/2 => 74 size after polling
#final size =>(74x74x16) => 16 is comning from filter size
#if none default its set to pool_size

#chunk2
model.add(tf.keras.layers.Conv2D(filters=32 ,kernel_size=(3,3), strides=(1,1), padding='valid', activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2), strides=None))
#final size => (32,32,32)

#FCNN
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(120,activation='relu'))
model.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```
model.summary()
# 3 channel(R,G,B) its has 16 filter. 9 values(3x3 matrix has 9 values random keral) in each filter
# (3 * 16 * 9) +16 bias => 448

#part2 
# as we have pool size 2 so its divides by 2 so 72.. polling will not have param

#part3
# conv2d_3 => 16 channnel(greyscale with 16 features) * 32(filter) * 9(matrix values) + bias(32) => 4640

#FNN
#41472*120 +bias => 4979790
#120*120 +bias => 14520
```

```
model.compile(optimizer=tf.keras.optimizers.SGD(),loss=tf.keras.losses.categorical_crossentropy,metrics=["acc"])
```

```
model= model.fit(train_gen, batch_size=32, epochs=1, validation_data = val_gen)
```

```
# get label Name   
def get_Label(number):
    labels = {0:'Cat', 1:'Dog'}
    return labels[number]


#plot predction function
def plot_predection(model_name):
    plt.figure(figsize=(20,15))
    plt.suptitle("Predection  Images", fontsize=20)
    images = [] 
    path ='/databricks/driver/test1/'
    count = 0  #val_images,val_labels
    for i,files in enumerate(os.listdir(path)) :
        img = plt.imread(path+files)
        img = cv2.resize(img,(120,120))
        plt.imshow(img,cmap=plt.cm.binary)
        img = np.expand_dims(img, axis=0)
        feature = model_name.predict(img)
        predection  = np.argmax(feature, axis=1)
        plt.subplot(5,7,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.xlabel("Predicted"+get_Label(int(predection)))
        count += 1
        if count == 34 :
            break 
```
